```python
import matplotlib.pyplot as plt
import pandas as pd

from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
df = pd.read_csv('2019.csv')
df.columns = ['Overall_rank','Country_or_region', 'Score', 'GDP_per_capitall', 
              'Social_support', 'Healthy_life_expectancy', 'Freedom_to_make_life_choices', 
              'Generosity', 'Perceptions_of_corruption']
df.head(15)


plt.plot(df.Overall_rank, df.Generosity)
plt.title('Зависимость степени щедрости населения от уровня счастья')
plt.xlabel('Место в рейтинге по уровню счастья')
plt.ylabel('Уровень щедрости')

#из получившегося графика видно, что уровень счастья населения никак не влияет на степень щедрости и наоборот
```

```python
df2 = df[['Overall_rank', 'GDP_per_capitall', 
              'Social_support', 'Healthy_life_expectancy', 'Freedom_to_make_life_choices', 
              'Generosity', 'Perceptions_of_corruption']].head(20)
df2.set_index('Overall_rank', inplace=True)
df2.head()

gh =  df2.plot(kind='area',
               stacked=False,
               title='Динамика показателей в зависимости от уровня счастья - топ 20',
               xlabel='Место в рейтинге по уровню счастья',
               ylabel='Значение показателя')
gh.legend(loc='upper right')

#видим, что самые равномерные значения у показателей социальной поддержки, ожидаемого срока продолжительности здоровой жизни и
#свободы выбора - из чего можем сделать вывод, что именно эти показатели оказывают существенное влияние на уровень счастья
#населения
```

```python
df3 = df[['Overall_rank', 'GDP_per_capitall', 
              'Social_support', 'Healthy_life_expectancy', 'Freedom_to_make_life_choices', 
              'Generosity', 'Perceptions_of_corruption']].tail(20)
df3.set_index('Overall_rank', inplace=True)
df3.head()

df3.plot(kind='bar', stacked=True, xlabel = 'Место в рейтинге по уровню счастья', ylabel = 'Значение показателя',
        title='Соотношение показателей в зависимости от уровня счастья - последние 20 стран рейтинга')

#из полученного графика видно, что большинство стран с низким уровнем счастья объединяет низкий уровень ВВП и низкие, иногда
#практически отсутствующие показатели значения свободы жизненного выбора
```

```python
df4 = df[['Overall_rank', 'GDP_per_capitall', 
              'Social_support', 'Healthy_life_expectancy', 'Freedom_to_make_life_choices', 
              'Generosity', 'Perceptions_of_corruption']]
df4.set_index('Overall_rank', inplace=True)

df4.Freedom_to_make_life_choices.plot(kind='box',
                   title='Распределение значений свободы жизненного выбора')

#данный график демонстрирует, что у большинства стран - участниц рейтинга оценка уровня свободы жизненного выбора ниже среднего
```
