```python
import pandas as pd

ratings = pd.read_csv('ratings.csv')
best_movieid = ratings.query('rating == 5.0').groupby('movieId').count().sort_values('rating', ascending=False).head(1)

movies = pd.read_csv('movies.csv')
joined = best_movieid.merge(movies, on='movieId', how='left')

print('Фильм, которому было выставлено больше всего оценок 5.0:')
joined.title.head()
```

```python
power = pd.read_csv('power.csv')

needed_countries = power[ (power['country'] == 'Latvia') | (power['country'] == 'Lithuania') | (power['country'] == 'Estonia') ]
country_category = needed_countries[ (needed_countries['category'] == 4) | (needed_countries['category'] == 12) | (needed_countries['category'] == 21) ]
power_sum = country_category.query('quantity >= 0').groupby('country').sum('quantity')

power_sum.quantity.head()
```
